```python

project_name = 'portraitxsdxl'
model_name = 'stabilityai/stable-diffusion-xl-base-1.0'
prompt = 'photo of a sdxlxyz man'
```

```python
print(model_name)
```

```python
# Inference
# this is the inference code that you can use after you have trained your model
# Unhide code below and change prj_path to your repo or local path (e.g. my_dreambooth_project)



from diffusers import DiffusionPipeline, StableDiffusionXLImg2ImgPipeline
import torch

prj_path = "./muhmodel/"
model = model_name
pipe = DiffusionPipeline.from_pretrained(
    model,
    torch_dtype=torch.float16,
)
pipe.to("cuda")
pipe.load_lora_weights(prj_path, weight_name="pytorch_lora_weights.safetensors")


neg_prompt = "|poorly rendered face, malformed face, extra fingers,poorly drawn hands, bad anatomy,\
               malformed limbs, fused fingers >-1:0"
```

```python
muh_prompt = prompt + ", a Pixar cartoon character, portrait "

muh_prompt += neg_prompt

seed = 42

for seed in range(10):
  generator = torch.Generator("cuda").manual_seed(seed)
  image = pipe(prompt=muh_prompt, generator=generator).images[0]
  # image = refiner(prompt=prompt, generator=generator, init_image =image).images[0]
  image.save("gen1_image"+str(seed * 3 + 1) +".png")
```
